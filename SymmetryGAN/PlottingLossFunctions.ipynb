```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm, ticker
%matplotlib inline
```

```python
def p(x,mu,sigma):
    return (1./np.sqrt(2*np.pi*sigma**2))*np.exp(-(x - mu)**2/(2.*sigma**2))
```

```python
def g(a, c, x):
    return a + c*x
def h(a, c, x):
    return p(x,0.5,1)/(p(x,0.5,1) + p(x,0.5*c+a,c))
    #return p(x,0.5,1)/(p(x,0.5,1) + abs(c)*p(a+c*x,0.5,1))
```

```python
xx = np.random.normal(0.5,1.,10000)
A = 0
C = 0
range = 1.5
alow = A-range
ahigh = A+range
astep = 100
clow = C - range
chigh = C + range
cstep = 100
alpha = 5

mylosses=[]
for c in np.linspace(clow,chigh,cstep):
    losses=[]
    for a in np.linspace(alow, ahigh, astep):
        yy = g(a, c, xx)
        losses += [np.mean(
                          - alpha*(g(a, c, yy)-xx)**2
                          )]
    mylosses += [losses]
mylosses = mylosses - np.array(mylosses).max()

x = np.linspace(alow, ahigh, astep)
y = np.linspace(clow, chigh, cstep)
X,Y = np.meshgrid(x,y)
Z = np.array(mylosses)


fig, ax = plt.subplots(figsize=(6,6))


ax.set_aspect('equal')
ax.set_xlabel('intercept')
ax.set_ylabel('slope')
cf = ax.contourf(X,Y,Z,50)
fig.colorbar(cf, ax=ax)
plt.scatter([0, 1], [1, -1], marker='*', c='blue')
plt.title('Loss')
ax.annotate('(0, 1)', xy=(0.1, 1))
ax.annotate('(1, -1)', xy=(1.1, -1))


plt.show()
```

```python
xx = np.random.normal(0.5,1.,10000)
A = 0
C = 1
range = 0.2
alow = A-range
ahigh = A+range
astep = 100
clow = C - range
chigh = C + range
cstep = 100

mylosses=[]
for c in np.linspace(clow,chigh,cstep):
    losses=[]
    for a in np.linspace(alow, ahigh, astep):
        yy = g(a, c, xx)
        losses += [np.mean(-np.log(0.000001+h(a, c, xx)) - np.log(0.000001+1.-h(a, c, yy))
                          - (g(a, c, yy)-xx)**2
                          )]
    mylosses += [losses]
mylosses = mylosses - np.array(mylosses).min() + 10**-5

x = np.linspace(alow, ahigh, astep)
y = np.linspace(clow, chigh, cstep)
X,Y = np.meshgrid(x,y)
Z = np.array(mylosses)


fig, ax = plt.subplots(figsize=(6,6))


ax.set_aspect('equal')
ax.set_xlabel('intercept')
ax.set_ylabel('slope')
cf = ax.contourf(X,Y,Z,30)
fig.colorbar(cf, ax=ax)
plt.scatter([0], [1], marker='*', c='yellow')
plt.title('g(x)  = x')


plt.show()
```
